````python
#1 
def bisection(f, l, u):
    tol = 0.0001
    max_iter = 1000
    fl = f(l)
    fu = f(u)
    assert fu*fl < 0, "f(u) & f(l) harus berubah tanda"
    for i in range(max_iter):
        r = (l+u)/2
        fr = f(r)
        if abs(fl*fr) < tol:
            return r, i
        if fl*fr < 0:
            u = r
            fu = fr
        else:
            l = r
            fl = fr


def f(i):
    #f(x)=0=(P * (i * (1+i)**n) / ((i+1)**n - 1) - A
    return (25000 * (i * (1+i)**6) / ((i+1)**6 - 1)) - 5500

root, iterasi = bisection(f, -0.5, 1)
print("Nomor 1 Bisection Method :")
print("Bunga: {:.4f}, Iterasi: {}".format(root, iterasi))

#output Akar: 0.0856, Iterasi: 20

#1.1 A = 7000 n = 5,5
def false_position(f, l, u):
  tol = 0.0001
  max_iter = 1000
  fl = f(l)
  fu = f(u)
  assert fu*fl < 0, "f(u) & f(l) harus berubah tanda"
  for i in range(max_iter):
        r = u - (fu * (l - u)) / (fl - fu)
        fr = f(r)
        if abs(fl*fr) < tol:
            return r, i
        if fl*fr < 0:
            u = r
            fu = fr
        else:
            l = r
            fl = fr

def f(i):
    return (25000 * (i * (1+i)**5.5) / ((i+1)**5.5 - 1)) - 7000

root, iterasi = false_position(f, -0.5, 1)
print("Nomor 1.1 false_position Method :")
print("Bunga: {:.4f}, Iterasi: {}".format(root, iterasi))

#Akar: 0.1505, Iterasi: 10

#1.2 P = 3000 i = 5% 
def bisection(f, l, u):
    tol = 0.0001
    max_iter = 1000
    fl = f(l)
    fu = f(u)
    assert fu*fl < 0, "f(u) & f(l) harus berubah tanda"
    for i in range(max_iter):
        r = (l+u)/2
        fr = f(r)
        if abs(fl*fr) < tol:
            return r, i
        if fl*fr < 0:
            u = r
            fu = fr
        else:
            l = r
            fl = fr
def f(A):
  return (3000 * (5 * (1+5)**6) / ((5+1)**6 - 1)) - A

root, iterasi = bisection(f, 0, 20000)
print("Nomor 1.2 Bisection Method :")
print("Biaya per bulan {:.4f}, Iterasi: {}".format(root, iterasi))

#Root 15000.3242, Iterasi: 19
````

FUNCTION newton_rhapson(f, x0, df)
    tol = 0.0000001
    max_iter = 1000
    fx0 = f(x0)
    dfx0 = df(x0)

    FOR i = 0 TO max_iter
        x1 = x0 - (fx0 / dfx0)
        error = ABS((x1 - x0) / x1) * 100
        IF error < tol THEN
            RETURN x1, i
        END IF
        x0 = x1
        fx0 = f(x0)
        dfx0 = df(x0)
        IF dfx0 == 0 THEN
            PRINT "Derivative is zero at x = {}. Newton-Rhapson method fails.".FORMAT(x0)
            RETURN None, None
        END IF
    END FOR
END FUNCTION

````python
#3
def newton_rhapson(f, x0, df):
  tol = 0.0000001
  max_iter = 1000
  fx0 = f(x0)
  dfx0 = df(x0)

  for i in range(max_iter):
        x1 =x0 - (fx0 / dfx0)
        error = abs((x1 - x0)/x1) * 100
        if error < tol:
            return x1, i
        x0 = x1
        fx0 = f(x0)
        dfx0 = df(x0)
        if dfx0 == 0:
             print("Derivative is zero at x = {}. Newton-Rhapson method fails.".format(x0))
             return None, None



def f(x):
  return (x / (1-x)) * math.sqrt(6/(2+x)) - 0.05

def df(x):
  return math.sqrt(6) * (x**2 + x + 4) / (2 * ((x-1)**2) * (x + 2)**(3/2) )

root, iterasi = newton_rhapson(f, 0, df)
print("x: {:.4f}, Iterasi: {}".format(root, iterasi))
````

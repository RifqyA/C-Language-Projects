```python
import math

def golden_section_search(f):
    tol=1e-6
    max_iter=100
    xold = 0
    xl = 200
    xu = 800
    # Calculate the golden ratio
    d = 0.61803 * (xu - xl)

    # Calculate the initial endpoints of the intervals
    x1 = xl + d
    x2 = xu - d

    # Iterate until convergence
    i = 0
    for i in range(max_iter):
        # Evaluate the function at the endpoints of the intervals
        f1 = round(f(x1),5)
        f2 = round(f(x2),5)
        print("iterasi = {}, x1 = {}, x2 = {}, f1 = {}, f2 = {}".format(i, round(x1,5) , round(x2, 5), f1, f2))
        # Choose the new interval

        if f1 > f2:
            xl = x2
            xnew = xl
        else:
            xu = x1
            xnew = xu

        if xold == 0:
          error = 100
        else:
          error = abs((xold - xnew) / xold)

        print("error = {}".format(error))
          #Return the value of x that minimizes/maximizes the function
        if error < tol:
            print("error = {}".format(error))
            return xnew, i
        xold = xnew

        # Calculate the new endpoints of the intervals
        d = 0.61803 * (xu - xl)
        x1 = xl + d
        x2 = xu - d

def f(x):
  return   0.01*0.6*x**2 + (0.95/0.6) * (16000/x)**2

xnew, iterasi = golden_section_search(f)
print("Hasil yang didapatkan adalah {} dengan iterasi {} Dengan f(X) {}".format(round(xnew, 5),iterasi,f(xnew)))
```
